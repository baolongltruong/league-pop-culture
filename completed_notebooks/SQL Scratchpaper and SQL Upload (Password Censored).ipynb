# Averaging the Vectors From Word2Vec

```python
comments = pd.read_csv("data/comments_dataframe.csv",index_col=[0])
```

```python
comments
```

```python
import gensim
from tqdm import tqdm
```

```python
model = gensim.models.Word2Vec.load("models/word2vec_5-window_2-min_10-epochs.model")
```

```python
avg_vector = list()
cid_list = list()
```

```python
vocabulary = model.wv.index_to_key
```

```python
for index, row in tqdm(comments.iterrows()):
    comment = row['comment_text']
    cid = row['comment_id']
    tokens = gensim.utils.simple_preprocess(comment)
    
    valid_vectors = list()
    
    for token in tokens:
        if token in vocabulary:
            valid_vectors.append(model.wv[token])
    np_vec = np.mean(valid_vectors, axis=0)
    list_vec = np_vec.tolist()
    cid_list.append(cid)
    avg_vector.append(list_vec)
```

```python
r = len(avg_vector)
```

```python
for i in range(r):
    if type(avg_vector[i]) is float:
        avg_vector[i] = [avg_vector[i]]
```

```python
i = 0
while i < r:
    if len(avg_vector[i]) < 100:
        avg_vector[i] = [None for n in range(100)]
    i += 1
```
